```python
import PyCESat2 as pysat
import numpy as np
```

```python
test_file = "../../../Kd_from_ATLAS/Data/Amsterdam/OpenAlt_Download/processed_ATL03_20190331065528_00310302_003_01.h5"
```

```python
atl = pysat.ATL03(test_file)
```

```python
t = atl.gt1l.RANSAC("R1")
```

```python
def lower(photon):
    photon = np.array([photon])
    model = t["R1"]["model"]
    return model.predict(photon.reshape(-1,1)) - 50

t = t.add_modeled_surface("L1",lower,t.distance)
```

```python
t = t.outside('R1','L1')
```
